```python
data['model'] = data['name'].str.split(' ').str[:3].str.join(' ')
data
```

```python
data.drop(['name'],axis=1,inplace=True)
data.head()
```

```python
#Checking all the comlumns
#Company: categorical
# year: int
# price: int
# kms: int
# fuel: categorical
# model: categorical
```

```python
data.info()
```

```python
data['year'].str.isnumeric().value_counts() #All true
data['year'] = data['year'].astype(int)
data.head()
```

```python
data['Price'] = data['Price'].str.replace(',','')
data.head()
```

```python
data = data[(data['Price'].str.isnumeric())].reset_index(drop=True)
data.head()
```

```python
data['Price'] =data['Price'].astype(int)
data.head()
```

```python
data['kms_driven'] = data['kms_driven'].str.split(' ').str[0]
data['kms_driven'] = data['kms_driven'].str.replace(',','')
data.head()
```

ENCODING

```python
from sklearn.preprocessing import OneHotEncoder
encoder_company = OneHotEncoder()
encoder_fuel = OneHotEncoder()
encoder_model = OneHotEncoder()

company = encoder_company.fit_transform(data[['company']]).toarray()
fuel = encoder_fuel.fit_transform(data[['fuel_type']]).toarray()
model = encoder_model.fit_transform(data[['model']]).toarray()
```

```python
df_company = pd.DataFrame(company,columns=encoder_company.get_feature_names_out())
df_fuel = pd.DataFrame(fuel,columns=encoder_fuel.get_feature_names_out())
df_model = pd.DataFrame(model,columns=encoder_model.get_feature_names_out())
```

```python
data_copy = data.copy()
```

```python
data_copy = pd.concat([data_copy,df_company,df_fuel,df_model],axis=1)
data_copy.head()
```

```python
data_copy.drop(['company','fuel_type','model'],axis=1,inplace=True)
data_copy.head()
```

```python
data_copy.shape
```

```python
data_copy.to_csv('clean_data.csv')
```

Outliers

```python
data = data_copy
```

```python
data.head()
```

```python
#Removing for price
q1 = data['Price'].quantile(0.25)
q3 = data['Price'].quantile(0.75)
IQR = q3-q1
l_f = q1 - 1.5*IQR
h_f = q3 + 1.5*IQR
data = data[(data['Price']>=l_f) & (data['Price']<=h_f)]
```

```python
#Removing for kms_driven
q1 = data['kms_driven'].quantile(0.25)
q3 = data['kms_driven'].quantile(0.75)
IQR = q3-q1
l_f = q1 - 1.5*IQR
h_f = q3 + 1.5*IQR
data = data[(data['kms_driven']>=l_f) & (data['kms_driven']<=h_f)]
```

```python
#Removing for year
q1 = data['year'].quantile(0.25)
q3 = data['year'].quantile(0.75)
IQR = q3-q1
l_f = q1 - 1.5*IQR
h_f = q3 + 1.5*IQR
data = data[(data['year']>=l_f) & (data['year']<=h_f)]
```

```python
data.to_csv('clean_data.csv',index=False)
```
